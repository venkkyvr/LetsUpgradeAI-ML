```python
crosstable = pd.crosstab(data.Attrition,data.BusinessTravel).apply(lambda r: r/r.sum(), axis=)
print(crosstable)

stats, p, dof, expected = chi2_contingency(crosstable)
print(f'The Value of p in chi2_contingency test is: {p}')
if p < 0.05:
    print('The variables are dependent')
else:
    print('The variables are not dependent')
```

```python
crosstable = pd.crosstab(data.Attrition,data.EducationField).apply(lambda r: r/r.sum(), axis=0)
print(crosstable)
stats, p, dof, expected = chi2_contingency(crosstable)
print(f'The Value of p in chi2_contingency test is: {p}')
if p < 0.05:
    print('The variables are dependent')
else:
    print('The variables are not dependent')
```
